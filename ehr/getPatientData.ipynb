```python
# relevant imports
import os
import pandas as pd
```

```python
def getPatientData(num, status, time):
    '''
    This function returns a specific patient's data at a specific time point
    
    :num: An integer that indicates patient number
    :status: A character that indicates 'R' for remission or 'F' for flare
    :time: The time point you wish to retreive
    :returns: a dataframe containing data for that specific patient
    '''
    
    df = pd.read_csv("index.csv")
    
    # get the location of the relevant data
    patient = 'Patient' + str(num)
    data_loc = df[(df['Sample'] == patient) & (df['Disease State'] == status) & (df['Time'] == time)]['Location']
    
    # check if this data exists
    if (data_loc.empty):
        print("This data does not exist")
        return()
    else:
        data_loc = data_loc.iloc[0]
 
    # get the relative path
    rel_pth = os.getcwd()
    fp = os.path.join(rel_pth, '..', '..', data_loc)
    fp = os.path.abspath(fp)                             # normalize the path
    
    data = pd.read_csv(fp)
    return(data)
    
```
